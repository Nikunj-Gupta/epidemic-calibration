```python
import pandas as pd 
```

```python
df = pd.read_csv("../data/ml_df.csv")
```

```python
df 
```

```python
def find_best(num_comp, amount_of_data_ranges, noise_ranges): 
    for (key_data, value_data) in amount_of_data_ranges.items(): 
        l_data, h_data = value_data 
        for (key_noise, value_noise) in noise_ranges.items(): 
            l_noise, h_noise = value_noise 

            print("amount_of_data_ranges:", key_data, l_data, h_data) 
            print("noise_ranges:", key_noise, l_noise, h_noise) 
            res = df[(df["num_compartments"]==num_comp) & 
                     (df["amount_of_data"]>=l_data) & 
                     (df["amount_of_data"]<=h_data) &
                     (df["noise"]<=h_noise) &
                     (df["noise"]>=l_noise)
                    ] 
            lst_1 = list(res['best_calibration_method']) 
            lst_2 = list(res['second_best_calibration_method']) 
            lst_3 = list(res['third_best_calibration_method']) 
            print("best_calibration_method on average:", max(lst_1, key=lst_1.count)) 
            print("second_best_calibration_method on average:", max(lst_2, key=lst_2.count)) 
            print("third_best_calibration_method on average:", max(lst_3, key=lst_3.count)) 
            print("================================================================") 
```

```python
amount_of_data_ranges = {
    "low": [27, 39], 
    "low_mid": [40, 49], 
    "mid_high": [50, 59], 
    "high": [60, 72],  
}

noise_ranges = {
    "low": [1, 5], 
    "mid": [6, 10], 
    "high": [11, 15], 
}
```

```python
find_best(
    num_comp=3, 
    amount_of_data_ranges=amount_of_data_ranges, 
    noise_ranges=noise_ranges
)
```

```python
find_best(
    num_comp=4, 
    amount_of_data_ranges=amount_of_data_ranges, 
    noise_ranges=noise_ranges
)
```
